```python
import pandas as pd
india_whether=pd.DataFrame({
    "city":['hydrabad','mumbai','Banglore','Delhi'],
    "tempreture":[36,38,29,39],
    "humidity":[56,65,89,44]
})
india_whether
```

```python
us_whether=pd.DataFrame({
    "city":['new york','toronto','chicago','orland'],
    "tempreture":[19,28,24,22],
    "humidity":[56,55,79,64]
})
us_whether
```

```python
df=pd.concat([india_whether,us_whether])
df
```

```python
df=pd.concat([india_whether,us_whether],ignore_index=True)
df
```

```python
df=pd.concat([india_whether,us_whether],keys=['india','us'])
df
```

```python
df.loc['india']
```

```python
df.loc['us']
```

```python
city_tempreture=pd.DataFrame({
    "city":['hydrabad','mumbai','Banglore','Delhi'],
    "tempreture":[36,38,29,39]
})
city_tempreture
```

```python
city_windspeed=pd.DataFrame({
    "city":['hydrabad','mumbai','Banglore','Delhi'],
    "windspeed":[6,8,9,7]
})
city_windspeed
```

```python
df=pd.concat([city_tempreture,city_windspeed],axis=1)
df
```

```python
city_tempreture=pd.DataFrame({
    "city":['hydrabad','mumbai','Banglore','Delhi'],
    "tempreture":[36,38,29,39]
},index=[0,1,2,3])
city_tempreture
```

```python
city_windspeed=pd.DataFrame({
    "city":['Banglore','mumbai','Delhi'],
    "windspeed":[8,9,7]
},index=[2,1,3])
city_windspeed
```

```python
df=pd.concat([city_tempreture,city_windspeed],axis=1)
df
```
