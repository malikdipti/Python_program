```python
import pandas as pd
df=pd.read_excel("servey.xls")
df
```

```python
pd.crosstab(df.Nationality,df.Handness)
```

```python
pd.crosstab(df.Sex,df.Handness)
```

```python
pd.crosstab(df.Sex,df.Handness,margins=True)
```

```python
pd.crosstab(df.Sex,[df.Nationality,df.Handness],margins=True)
```

```python
pd.crosstab([df.Sex,df.Nationality],[df.Handness],margins=True)
```

```python
pd.crosstab([df.Sex],[df.Handness],normalize="index")
```

```python
import numpy as np
pd.crosstab([df.Sex],[df.Handness],values=df.Age,aggfunc=np.average)
```
