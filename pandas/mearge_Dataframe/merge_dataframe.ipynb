```python
import pandas as pd
df1=pd.DataFrame({'city':['hyd','bbsr','banglore'],
                  'tempreture':[32,34,28]
})
df1
```

```python
df2=pd.DataFrame({'city':['hyd','bbsr','banglore'],
                  'humidity':[78,64,58]
})
df2
```

```python
df3=pd.merge(df1,df2,on='city')
df3
```

```python
df1=pd.DataFrame({'city':['hyd','bbsr','banglore','mumbai'],
                  'tempreture':[32,34,28,29]
})
df1
```

```python
df2=pd.DataFrame({'city':['hyd','bbsr','delhi'],
                  'humidity':[78,64,58]
})
df2
```

```python
df3=pd.merge(df1,df2,on="city")
df3
```

```python
df3=pd.merge(df1,df2,on="city",how='outer')
df3
```

```python
df3=pd.merge(df1,df2,on="city",how='left')
df3 
```

```python
df3=pd.merge(df1,df2,on="city",how='right')
df3
```

```python
df3=pd.merge(df1,df2,on="city",how='outer',indicator=True)
df3
```

```python
df1=pd.DataFrame({'city':['hyd','bbsr','banglore','mumbai'],
                  'tempreture':[32,34,28,29],
                  'humidity':[89,96,58,68]
})
df1
```

```python
df2=pd.DataFrame({'city':['hyd','bbsr','chicago'],
                  'tempreture':[32,34,28],
                  'humidity':[89,96,68]
})
df2
```

```python
df3=pd.merge(df1,df2,on='city',suffixes=('_left','_right'))
df3
```
