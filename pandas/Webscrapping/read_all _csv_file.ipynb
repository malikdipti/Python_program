```python
import pandas as pd
import glob

path = r'D:\python project\pandas\Webscrapping' # use your path
all_files = glob.glob(path + "/*.csv")

li = []

for filename in all_files:
    df = pd.read_csv(filename, index_col=None, header=0)
    li.append(df)

frame = pd.concat(li, axis=0, sort=False)
frame
```
